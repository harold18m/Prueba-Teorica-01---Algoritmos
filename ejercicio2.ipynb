```python
"""
Se tiene un texto extremadamente largo, aplicar algoritmo de fuerza bruta para buscar
10 palabras contenidas en una lista. e indicar el tiempo de demora de cada elemento en
la serie, mostrando con un gráfico donde:
x=palabras; y= tiempo de ejecución (5 puntos)
"""

import time
import matplotlib.pyplot as plt

texto = "Este es un texto extremadamente largo que contiene muchas palabras interesantes. Queremos buscar algunas de estas palabras utilizando el algoritmo de fuerza bruta."

palabras = ["texto", "palabras", "algoritmo", "fuerza", "bruta", "largo", "buscar", "interesantes", "muchas", "ninguna"]

tiempos = []

for palabra in palabras:
    inicio = time.time()
    if texto.find(palabra) != -1:
        print("La palabra", palabra, "se encontró en el texto.")
    else:
        print("La palabra", palabra, "no se encontró en el texto.")
    fin = time.time()
    tiempo = fin - inicio
    tiempos.append(tiempo)
    print("Tiempo de búsqueda:", tiempo, "segundos\n")

plt.bar(palabras, tiempos)
plt.xlabel("Palabras")
plt.ylabel("Tiempo de ejecución (segundos)")
plt.title("Tiempo de ejecución de búsqueda de palabras")
plt.show()
```
